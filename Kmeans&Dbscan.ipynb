```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.cluster import KMeans
from sklearn.datasets import make_moons
from sklearn.datasets import make_blobs
```

```python
X, _ = make_moons(n_samples=100)
```

```python
plt.scatter(X[:,1], X[:,0])
plt.show()
#before clustering
```

```python
kmeans = KMeans(n_clusters = 2)
results = kmeans.fit_predict(X)

plt.scatter(X[:,1], X[:,0], c = results, cmap = 'coolwarm')
plt.show()
#KMeans
```

```python
list1=[]
list2=[]

for i in range(len(results)):
  if (results[i] == 1):
    list1.append(X[i])
  else:
    list2.append(X[i])
```

```python
#results = results.reshape(-1 ,1)
dbscan1 = DBSCAN(eps = 0.2, min_samples = 2)
dbscan1.fit(list1)
dbscan1.labels_
```

```python
dbscan2 = DBSCAN(eps = 0.2, min_samples = 2)
dbscan2.fit(list2)
dbscan2.labels_
```

```python
list1 = np.array(list1)
list2 = np.array(list2)
```

```python
#res = DBSCAN.fit_predict(X,0)
plt.scatter(list1[:,1], list1[:,0], c=dbscan1.labels_)
plt.show()
#DBSCAN
```

```python
plt.scatter(list2[:,1], list2[:,0], c=dbscan2.labels_)
plt.show()
```

```python
plt.scatter(list1[:,1], list1[:,0], c=dbscan1.labels_)
plt.scatter(list2[:,1], list2[:,0], c=dbscan2.labels_)
plt.show()
```

```python
X, _ = make_blobs(n_samples=1000, centers=3, n_features=2,random_state=0, cluster_std=0.55)
plt.scatter(X[:,0], X[:,1])
plt.show()
#no clustering
```

```python
kmeans = KMeans(n_clusters=2)
results = kmeans.fit_predict(X)
plt.scatter(X[:,0], X[:,1], c=results, cmap="coolwarm")
plt.show()
#using KMeans
```

```python
list3 = []
list4 = []

for i in range(len(results)):
  if (results[i] == 1):
    list3.append(X[i])
  else:
    list4.append(X[i])
```

```python
dbscan3 = DBSCAN(eps = 0.2, min_samples = 3)
dbscan3.fit(list3)
dbscan3.labels_
```

```python
dbscan4 = DBSCAN(eps = 0.2, min_samples = 3)
dbscan4.fit(list4)
dbscan4.labels_
```

```python
list3 = np.array(list3)
list4 = np.array(list4)
```

```python
#res = DBSCAN.fit_predict(X,0)
plt.scatter(list3[:,1], list3[:,0], c=dbscan3.labels_)
plt.show()
#DBSCAN
```

```python
#res = DBSCAN.fit_predict(X,0)
plt.scatter(list4[:,1], list4[:,0], c=dbscan4.labels_)
plt.show()
#DBSCAN
```
